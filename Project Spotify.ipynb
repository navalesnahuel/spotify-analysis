# Rank artists by Streams

```python
artist_streams = df.groupby('artist(s)_name')['streams'].sum().sort_values(ascending=False).head(20)

# Plot a horizontal bar plot for the top N artists
ax = artist_streams[::-1].plot(kind='barh', legend=False)

plt.xlabel('Total Streams')
plt.ylabel('Artist(s) Name')
plt.title('Top 10 Artists by Total Streams')
plt.show()
```

# Songs by BPM

```python
top_dance = df.groupby("track_name")["bpm"].mean().sort_values(ascending=False).head(20)
ax = top_dance[::-1].plot(kind="barh", legend=False)
plt.xlabel('BPM')
plt.ylabel('Track Name')
plt.title("Songs by BPM")
plt.show()
```

# Most Danceable Songs

```python
positive_songs = df.groupby("track_name")["danceability_%"].mean().sort_values(ascending=False).head(20)
ax = positive_songs[::-1].plot(kind="barh", legend=False)
plt.xlabel('Danceability')
plt.ylabel('Track Name')
plt.title("Most danceable songs")
plt.show()
```

# Songs by Valence

```python
positive_songs = df.groupby("track_name")["valence_%"].mean().sort_values(ascending=False).head(20)
ax = positive_songs[::-1].plot(kind="barh", legend=False)
plt.xlabel('Valence')
plt.ylabel('Track Name')
plt.title("Top 20 Valence songs")
plt.show()
```

# Songs by Energy

```python
positive_songs = df.groupby("track_name")["energy_%"].mean().sort_values(ascending=False).head(20)
ax = positive_songs[::-1].plot(kind="barh", legend=False)
plt.xlabel('Energy')
plt.ylabel('Track Name')
plt.title("Top 20 Most energic songs")
plt.show()
```

# Songs by quality

```python
top_tracks_data = df.groupby('track_name')[['danceability_%', 'valence_%', 'energy_%']].mean()
top_tracks_data = top_tracks_data.mean(axis=1).sort_values(ascending=False).head(20)

# Plot a grouped bar plot
ax = top_tracks_data[::-1].plot(kind='barh')
plt.xlabel('Track Name')
plt.ylabel('Percentage / Total Streams')
plt.title('Top 20 Songs by Danceability %, Valence %, and Energy %')
plt.show()
```

# Most reproduced Songs

```python
songs_in_playlist = df.groupby("track_name")["in_spotify_playlists"].mean().sort_values(ascending=False).head(20)
ax = songs_in_playlist[::-1].plot(kind="barh", legend=False)
plt.xlabel('Playlist')
plt.ylabel('Track Name')
plt.title("Top 20 most reproduced songs")
plt.show()
```

# The king of the Playlists

```python
king_of_playlists = df.groupby("artist(s)_name")["in_spotify_playlists"].sum().sort_values(ascending=False).head(5)
ax = king_of_playlists[::-1].plot(kind="barh", legend=False, color="yellow")
plt.xlabel("Playlist ammount")
plt.ylabel("Artists")
plt.title("King of spotify playlists")
plt.show()
```
